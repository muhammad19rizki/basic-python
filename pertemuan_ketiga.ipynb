```python
password = "secret"
passwordInput = ""
count = 0
max = 5
auth = False

while passwordInput != password:
    auth = True
    count += 1
    if count > max:
        auth = False 
        break
    passwordInput = input("Masukkan Password : ")

if auth:
    print("Okey")
else:
    print("Call Police")
```

```python
def prima(n):
    if n <= 1:
        return False
    for x in range(2, n):
        if n % x == 0:
            return False
    else:
        return True 
```

```python
n = 4
if prima(n):
    print(f"{n} adalah bilangan prima")
else:
    print(f"{n} bukan bilangan prima")
```

```python
def bilanganPrima():
    for n in range(100):
        if prima(n):
            print(n, end="-")
    print()
bilanganPrima()
```

```python
def main():
        x = dict(pagi= "Kentang Goreng", siang = "Rendang", malam = "Geprek")
        makanan(**x)

def makanan(**kwargs):
    if len(kwargs):
        for k in kwargs:
            print(f"Sarapan {k} saya adalah {kwargs[k]}")
    else: pass

main()
```
